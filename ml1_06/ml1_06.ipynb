## **問題１**
微分の定義は以下であった。この数値微分は、$h$の値を微小にすることで、計算が可能である。
$$f'(x) = \lim_{h\rightarrow 0}\frac{f(x+h)-f(h)}{h}　\longrightarrow f'(x) =\frac{f(x+h)-f(h)}{h}$$
この数値微分の定義は、前方差分と呼ばれる。また、誤差を減らす工夫として、以下のような中心差分が利用される。
$$f'(x) =\frac{f(x+h)-f(x-h)}{2h}$$

$$f(x,y)=x^2 + y^2$$

$$L(a,b)=\sum_{i=1}^n(ax_i+b-y_i)^2$$
