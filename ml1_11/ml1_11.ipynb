## **問題１**
ある学校の生徒が学校で表彰された回数（awards）を、生徒の数学の試験の点数（math）と生徒の選んだ進路（acは大学進学、voは職業訓練校進学）を用いて予測したい。架空のデータ`data/data01.csv`に対してポアソン回帰を行うプログラムを`ml1_11_01.py`として作成した。このデータに対して、ポアソン回帰を行うことで、以下の結果に対する解釈を完成させよ。ただし、小数点第２位までの数値として求めること。
- 他の説明変数の値を固定した際に、数学の試験の点数が1点上昇すると、 生徒が表彰される回数は〇〇倍増加する。
- 他の説明変数の値を固定した際に、大学に進学した生徒は、 表彰された回数が〇〇倍増加する。

ただし、必要に応じて以下を参考にすること。ポアソン回帰モデルは以下である。
$$y=\exp(w_0+w_1x_1+\cdots+w_dx_d )$$

$$\frac{\exp(w_0+w_1x_1+\cdots+w_i(x_i+1)+\cdots+w_dx_d )}{\exp(w_0+w_1x_1++\cdots+w_ix_i+\cdots+w_dx_d )}=\exp(w_i)$$
